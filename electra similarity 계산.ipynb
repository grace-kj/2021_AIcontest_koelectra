```python
train_df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/국립국어원_Electra/data/BoolQ/SKT_BoolQ_Train.tsv', delimiter= '\t')
train_df
```

```python
train_df['Text list']= np.nan
train_df['Text list']= train_df['Text'].apply(lambda x : kss.split_sentences(x))
train_df['Text list'][0]
```

```python
train_df['similarity'] = [[] for _ in range(len(train_df))]
train_df['calculated Text'] = ['' for _ in range(len(train_df))]
train_df
```

```python
from transformers import ElectraModel, ElectraTokenizer

cos = torch.nn.CosineSimilarity()
tokenizer = ElectraTokenizer.from_pretrained('monologg/koelectra-base-v3-discriminator')
model = ElectraModel.from_pretrained('monologg/koelectra-base-v3-discriminator')
```

```python
model.cuda()
```

```python
for index, sen_list in tqdm(enumerate(train_df['Text list'])):
  ret_list=[]

  #text list의 문장들에 대해서
  for num, sentence in enumerate(sen_list):
    #similarity 계산
    text1 = train_df['Question'][index]
    text2 = sentence

    inputs_1 = tokenizer(text1, return_tensors='pt')
    inputs_2 = tokenizer(text2, return_tensors='pt')

    inputs_1 = {k:v.cuda() for k, v in inputs_1.items()}
    inputs_2 = {k:v.cuda() for k, v in inputs_2.items()}
#batch = tuple(t.to(args.device) for t in batch)

    cls1 = model(**inputs_1)[0][:, 0]
    cls2 = model(**inputs_2)[0][:, 0]

    cosine_similarity = cos(cls1, cls2).item()

    #cos similarity 추가
    ret_list.append([num,cosine_similarity])

  train_df['similarity'][index]= ret_list

train_df
```

```python
train_df.to_csv('BoolQ_similarity_Train.tsv', index= False, sep='\t')
```

```python
#########################count 바꾸기 위해 similarity 파일 불러올 경우 스트링 리스트 처리
train_df= pd.read_csv('BoolQ_similarity_Train.tsv', sep='\t')

#ret_list= [[],[],,,,[]]
ret_list=[[] for _ in range(len(train_df))]
for index, str1 in enumerate(train_df['similarity']):
  numbers=''

  #str1[1:-1]= [0, 0.9876], [1, 0.9876]
  for char in str1[1:-1]:
    if char=='[' or char== ']':
      pass
    else:
      numbers+=char

  num_list= list(map(float, numbers.split(', ')))

  for i in range(int(len(num_list)/2)):
    sim_list= num_list[2*i:2*i+2]
    sim_list[0]=int(sim_list[0])
    ret_list[index].append(sim_list)

train_df['similarity']=ret_list

train_df['Text list']= train_df['Text'].apply(lambda x : x.split('.'))
train_df

############################
```

```python
#몇 개 뽑을지
count=4
#train_df 전체 길이만큼 돌면서 similarity
for index, sim_list in enumerate(train_df['similarity']):
  sim_list.sort(key= lambda x: x[1], reverse= True)

  #가장 similarity 높은 것 뽑음
  if len(sim_list)>=count:
    sim_list=sim_list[:count]

  ret_sentence= ''
  sim_list.sort()
  for item in sim_list:
    num= item[0]
    ret_sentence= ret_sentence + train_df['Text list'][index][num] + '.'

  train_df['calculated Text'][index]= ret_sentence

train_df
```

```python
df= train_df[['ID','calculated Text','Question','Answer(FALSE = 0, TRUE = 1)']]
df.columns=['ID','Text','Question','Answer(FALSE = 0, TRUE = 1)']
df
```

```python
df.to_csv('BoolQ_Sim4_Train.tsv', index= False, sep='\t')
```
